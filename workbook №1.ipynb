```python
from matplotlib import pyplot as plt
import math

function = []
for x in range(1,11): 
      function.append((math.sqrt(1 + math.e ** math.sqrt(x) + math.cos(x**2)))/abs(1- (math.sin(x))**3)) 
newfunction = function[:5] 
plt.plot(function) 
plt.show()
fig, ax = plt.subplots(figsize=(10, 6)) 
ax.scatter(newfunction, newfunction)
plt.show()
```

```python
from numpy import trapz
import numpy as np

x = np.arange(0.0, 10)  
y = np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x+1)))) 

plt.grid()
plt.plot(x, y, c = 'r')
plt.fill_between(x,y)
plt.show()

print(trapz(y)) 
```

```python
from math import *

x = int(input('Введите x: '))
y = int(input('Введите y:'))
action = int(input('Введите действие (1 - +, 2 - -, 3 - *, 4 - /, 5 - e^(x=y), 6 - sin(x+y), 7 - cos(x+y), 8 - x**y):'))

if action == 1:
    print(x+y)
elif action == 2:
    print(x-y)
elif action == 3:
    print(x*y)
elif action == 4:
    if (y == 0):
        print('На ноль делить нельзя!')
    else:
        print(x/y)
elif action == 5:
    print(e**(x+y))
elif action == 6:
    print(sin(x+y))
elif action == 7:
    print(cos(x+y))
elif action == 8:
    print(x**y)
```
